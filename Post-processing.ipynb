```python
import vtk
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.linalg as sl
from pprint import pprint
from vtk.util.numpy_support import vtk_to_numpy
```

```python
def vtk_2_DataF(filename):
    reader = vtk.vtkXMLPolyDataReader()
    reader.SetFileName(filename)
    reader.Update()
    pdata = reader.GetOutput()
    Velocity_x, Velocity_y,Velocity, Position_x, Position_y, Pressure = [], [], [], [], [], []
    for i in range(pdata.GetNumberOfPoints()):
        Pressure.append(pdata.GetPointData().GetArray('Pressure').GetValue(i))
        Position_x.append(pdata.GetPoint(i)[0])
        Position_y.append(pdata.GetPoint(i)[1])
        
        
        Velocity_x.append(pdata.GetPointData().GetArray('Velocity').GetTuple(i)[0])
        Velocity.append(np.sqrt((np.square(pdata.GetPointData().GetArray('Velocity').GetTuple(i)[0]) +
                   np.square(pdata.GetPointData().GetArray('Velocity').GetTuple(i)[0]))))

        Velocity_y.append(pdata.GetPointData().GetArray('Velocity').GetTuple(i)[1])
        df = pd.DataFrame({
            "Position_x":Position_x , "Position_y": Position_y, "Pressure": Pressure, "Velocity_x": Velocity_x, "Velocity_y": Velocity_y
  , "Velocity":Velocity})
        
        
        df = df.loc[(df['Position_x'] > 0) & (df['Position_x'] < 20) & (df['Position_y'] > 0) & (df['Position_y'] < 10)]
        
#         df.sort_values('Position_y', ascending=False, inplace = True)
        
        df.reset_index(drop=True, inplace=True)
        
        
    return df
```

```python
each_step_x = []
each_step_y = []
Direction = 0.
t = 0.
time = []
for i in range (0,299):
    print(i)
    path = "/Users/hezhu/Desktop/tests010/step_"+str(i)+".vtp" 
    
    p = vtk_2_DataF(path)
    print(p.shape)

    a = 0.
    b = 0.
    c = 0.

    c = p[p.Velocity_x == 0].size
    
    a = p[p.Velocity_x > 0].size
    
    b = p[p.Velocity_x < 0].size
    
            
    x_available = []
    y_available = []

    if((c > a ) and (c > b)):
        new_dataf_1 = p.nlargest(1, 'Position_y')
        new_dataf_1.reset_index(drop=True, inplace=True)
        
        x_available.append((new_dataf_1.loc[0].Position_x))
        y_available.append(new_dataf_1.loc[0].Position_y)

    else:
        new_dataf_2 = p.nlargest(10, 'Velocity_x')
        new_dataf_2.reset_index(drop=True, inplace=True)

        if((a > c ) and (a > b)):
            Direction = 1
        if((b > a ) and (b > c)):
            Direction = -1
        for k in range(10):
            if(new_dataf_2.loc[k].Velocity_x / Direction > 0):
                x_available.append(new_dataf_2.loc[k].Position_x)

                y_available.append(new_dataf_2.loc[k].Position_y)



    each_step_x.append(np.median(x_available))

    each_step_y.append(np.median(y_available))

#     t += 0.01
#     time.append(t)

time = np.linspace(0,300)
    # p = vtk_2_DataF("/Users/hezhu/Desktop/post-processing/tests0.vtp")
    # vtk_2_DataF("/Users/hezhu/Desktop/post-processing/tests0.vtp")
```
